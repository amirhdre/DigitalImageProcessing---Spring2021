```python
image = cv.imread("noisy_image.png", flags=cv.IMREAD_GRAYSCALE)
fig, ax = plt.subplots(figsize=(5,5))
st = fig.suptitle("Chest Image", fontsize="x-large")
ax.imshow(image,vmin=image.min(),vmax=image.max(), cmap='gray')
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
def normalize(s):
    start = 0.0
    end = 255.0
    width = end - start
    res = (s - s.min())/(s.max() - s.min()) * width + start
    return res.astype(np.uint8)

# Take the 2-dimensional DFT and centre the frequencies
ftimage = cv.dft(np.float32(image),flags = cv.DFT_COMPLEX_OUTPUT)
ftimage = np.fft.fftshift(ftimage)
ftimage = normalize(20*np.log(cv.magnitude(ftimage[:,:,0],ftimage[:,:,1])))
fig, ax = plt.subplots(figsize=(5,5))
st = fig.suptitle("DFT of Chest Image", fontsize="x-large")
ax.imshow(np.abs(ftimage),vmin=ftimage.min(),vmax=ftimage.max(), cmap='gray')
ax.axis('off')
plt.show()
```

```python
img = np.zeros((415,415))

x, y = np.indices(img.shape)
r, c = img.shape
dis = (x-r//2)**2+(y-c//2)**2
img[(dis < 155**2) ^ (dis < 150**2)] = 0.2 * img.size
# img[(dis < 101**2) ^ (dis < 100**2)] = 0.5 * img.size
# img[(dis < 151**2) ^ (dis < 150**2)] = 0.2 * img.size
# img[(dis < 201**2) ^ (dis < 200**2)] = 0.4 * img.size
# img[(dis < 251**2) ^ (dis < 250**2)] = 0.8 * img.size

img = 1 - img

fig, ax = plt.subplots(figsize=(5,5))
st = fig.suptitle("Frequency Filter", fontsize="x-large")
ax.imshow(img,vmin=img.min(),vmax=img.max(), cmap='gray')
ax.axis('off')
plt.show()
```

```python
image_dft = np.fft.fft2(image)
image_dft_shift = np.fft.fftshift(image_dft)
image_dft_shift_ms = normalize(np.log(np.abs(image_dft_shift)))

filtered = np.copy(image_dft_shift)
x, y = np.indices((415,415))
r, c = (415,415)
dis = (x-r//2)**2+(y-c//2)**2
filtered[(dis < 160**2) ^ (dis < 140**2)] = 255
filtered_ms = normalize(np.log(np.abs(filtered)))

# reversing image
f_ishift = np.fft.ifftshift(filtered)
img_back = np.fft.ifft2(f_ishift)
img_back = np.abs(img_back)

fig, ax = plt.subplots(figsize=(10,10))
ax.imshow(img_back, cmap='gray')
ax.axis('off')
plt.show()
```
