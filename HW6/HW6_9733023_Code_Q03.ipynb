```python
image = cv.imread('CBC.jpg', flags=cv.IMREAD_GRAYSCALE)
fig, ax = plt.subplots(figsize=(8,8))
st = fig.suptitle("Original Image", fontsize="x-large", y=.85)
ax.imshow(image,vmin=image.min(),vmax=image.max(), cmap='gray')
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
d1 = 20
d2 = 25
edges = cv.Canny(image, d1, d2)
fig, ax = plt.subplots(figsize=(8,8))
st = fig.suptitle("Edges Image", fontsize="x-large", y=.85)
ax.imshow(edges,vmin=edges.min(),vmax=edges.max(), cmap='gray')
ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
image = cv.imread('CBC.jpg', flags=cv.IMREAD_GRAYSCALE)
cimg = cv.cvtColor(image,cv.COLOR_GRAY2BGR)
bimage = cv.medianBlur(image,5)
circles = cv.HoughCircles(bimage,method=cv.HOUGH_GRADIENT,dp=1,minDist=100, \
                            param1=27,param2=30,minRadius=80,maxRadius=120)
circles = np.uint16(np.around(circles))

for i in circles[0,:]:
    # draw the outer circle
    cv.circle(cimg,(i[0],i[1]),i[2],(0,255,0),6)
    # draw the center of the circle
    cv.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
    
fig, ax = plt.subplots(figsize=(8,8))
st = fig.suptitle("detected circles", fontsize="x-large", y=.85)
ax.imshow(cimg)
ax.axis('off')
plt.tight_layout()
plt.show()
```
